## Generate Single Image
Use one of the following emotions as input: fear, neutral, disgust, angry, surprised, sad, happy, or random.

*Note: At the first execution the StyleGan3 model will be downloaded, this may take some time.*

````python
img = gen_emotion('happy', 'Happy Person')
````

You can also adjust the 'degree' of the emotion using the random_factor argument.

Reducing the randomness will increase the emotion BUT make resulting images more alike.

````python
img_2 = img = gen_emotion('happy', 'Very Happy Person', random_factor=0.15)
````
