```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate a test image with a sinusoidal pattern.
# Adjust amplitude, frequency, and phase as needed.
amplitude = 1
frequency = 0.5  # Change this frequency for different patterns.
phase = 0

# Create a grid of X and Y coordinates.
x, y = np.meshgrid(np.arange(-10, 10, 0.1), np.arange(-10, 10, 0.1))

# Generate the sinusoidal pattern.
pattern = amplitude * np.sin(2 * np.pi * frequency * x + phase)

# Apply the optical system transformation (MTF) to the pattern.
# Here we assume a simple MTF for demonstration.
mtf = np.exp(-0.5 * (x**2 + y**2))

# Apply MTF to the pattern.
output_pattern = pattern * mtf

# Visualize the input and output patterns.
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(pattern, cmap='gray')
plt.title('Input Pattern')
plt.subplot(1, 2, 2)
plt.imshow(output_pattern, cmap='gray')
plt.title('Output Pattern with Applied MTF')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a range of spatial frequencies (cycles/degree).
spatial_frequencies = np.arange(0.1, 30.1, 0.1)

# Define the Contrast Sensitivity Function (CSF).
# This is just an example; you could use a more complex function based on real research.
csf = np.exp(-0.5 * (np.log(spatial_frequencies) - np.log(5))**2 / (0.25**2))

# Find the frequency of maximum sensitivity.
max_csf = np.max(csf)
max_spatial_frequency = spatial_frequencies[np.argmax(csf)]

# Visualize the CSF.
plt.plot(spatial_frequencies, csf, 'b')
plt.plot(max_spatial_frequency, max_csf, 'ro')
plt.xlabel('Spatial Frequency (cycles/degree)')
plt.ylabel('Contrast Sensitivity')
plt.title('Contrast Sensitivity Function (CSF)')
plt.legend(['CSF', 'Maximum Sensitivity'])
plt.show()
```
